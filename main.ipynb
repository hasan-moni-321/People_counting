```python
import numpy as np

import cvzone
import cv2

from ultralytics import YOLO
```

```python
# Function for finding the "Area of interest"
def RGB(event, x, y, flags, param):
    if event == cv2.EVENT_MOUSEMOVE:
        point = [x, y]
        print(point)

cv2.namedWindow('RGB')
cv2.setMouseCallback('RGB', RGB)

# Download the YOLO11s YOLO_model for detection
YOLO_model = YOLO("yolo11s.pt")
names = YOLO_model.model.names

# Read the given Vido using VideoCapture of opencv 
video_capture = cv2.VideoCapture('/home/hasan/CultureHint_Assignment/Computer Vision task Culture Hint.mp4')
count=0
area_of_interest_line_1 = 745 #506
area_of_interest_line_2 = 770 #531
offset_value = 8
input_value={}
enter_value= []
exit_value={}
out_value=[]
while True:
    ret,frame = video_capture.read()
    if not ret:
        break
    count += 1
    if count % 3 != 0:
        continue

    frame = cv2.resize(frame, (1620, 900))
    
    # Run YOLO11 tracking on the frame, persisting tracks between frames
    results = YOLO_model.track(frame, persist=True,classes=0)

    # Check if there are any boxes in the results
    if results[0].boxes is not None and results[0].boxes.id is not None:
        # Get the boxes (x, y, w, h), class IDs, track IDs, and confidences
        boxes = results[0].boxes.xyxy.int().cpu().tolist()  # Bounding boxes
        class_ids = results[0].boxes.cls.int().cpu().tolist()  # Class IDs
        track_ids = results[0].boxes.id.int().cpu().tolist()  # Track IDs
        confidences = results[0].boxes.conf.cpu().tolist()  # Confidence score
       
        for box, class_id, track_id, conf in zip(boxes, class_ids, track_ids, confidences):
            c = names[class_id]
            x1, y1, x2, y2 = box
            cx = int(x1+x2)//2 
            cy = int(y1+y2)//2 

            ############################################################################
            # Entering Part
            # Counting and detecting entering people (people coming to the bottom of the video)
            if area_of_interest_line_1<(cy+offset_value) and area_of_interest_line_1>(cy-offset_value): 
                input_value[track_id] = (cx, cy) 
            if track_id in input_value: 
                if area_of_interest_line_2 < (cy+offset_value) and area_of_interest_line_2 > (cy - offset_value): 
                    cv2.circle(frame, (cx, cy), 4, (255, 0, 0), -1) # circle in the middle of the bounding box
                    cv2.rectangle(frame, (x1, y1),(x2, y2),(255, 0, 0), 2) # rectangle for detecting human when in the area of interest(line)
                    cvzone.putTextRect(frame,f'{track_id}',(x1,y2),1,1) # for track_id visualization 
                    cvzone.putTextRect(frame,f'{c}',(x1,y1),1,1) # 
                    if enter_value. count(track_id) == 0: 
                        enter_value. append(track_id) 
                  
            ####################################################################
            # Exiting Part
            # Counting and detecting exiting people (people coming from bottom to the top of the video) 
            if area_of_interest_line_2<(cy+offset_value) and area_of_interest_line_2>(cy-offset_value): 
                exit_value[track_id] = (cx, cy) 
            if track_id in exit_value:  
                if area_of_interest_line_1 < (cy+offset_value) and area_of_interest_line_1 > (cy - offset_value): 

                    cv2.circle(frame, (cx, cy), 4, (255, 0, 0), -1) 
                    cv2.rectangle(frame, (x1, y1),(x2, y2),(255, 0, 0), 2)
                    cvzone.putTextRect(frame,f'{track_id}',(x1,y2),1,1)
                    cvzone.putTextRect(frame,f'{c}',(x1,y1),1,1)
                    if out_value.count(track_id) == 0: 
                        out_value.append(track_id) 

    # Draw two line for area of interest for entering and exiting people. 
    #cv2.line(frame,(480,531),(795,531),(255,0,255),2)
    #cv2.line(frame,(480,506),(785,506),(0,0,255),2)

    cv2.line(frame,(750,770), (1250, 770), (255,0,255), 2) 
    cv2.line(frame,(750,745), (1234, 745), (0,0,255), 2) 
    
    # Counting total exiting and entering people 
    enterp = len(enter_value)    
    exitp = len(out_value) 
    # Displaying total entering and exiting people  
    cvzone.putTextRect(frame, f'EnterPerson: {enterp}', (50, 60), 2, 2)
    cvzone.putTextRect(frame, f'ExitPerson: {exitp}', (50, 160), 2, 2)

    # Console print 
    print("Total number of people entering: ", enterp)
    print("Total number of people exiting: ", exitp) 

    # For visualizing frame and waitkey for exiting when press "q"  
    cv2.imshow("RGB", frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
       break

# Release the video capture object and close the display window
video_capture.release()
cv2.destroyAllWindows()

```
